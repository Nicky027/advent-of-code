## Part 1

```python
file = open('code.txt','r')
```

```python
code = file.read()
```

```python
code = code.replace('\n','')
```

```python
code_array = code.split(',')
```

```python
code_array = [int(x) for x in code_array]
```

```python
code_array[1] = 12
code_array[2] = 2
```

```python
def MakeCalculation(input_array):
    code_array = input_array

    i = 0 

    while i < len(code_array):
        if code_array[i] == 99:
            print('Calculation Done.')
            break
        elif code_array[i] == 1:
            j1 = code_array[i+1]
            j2 = code_array[i+2]
            j3 = code_array[i+3]
            code_array[j3] = code_array[j1] + code_array[j2]
            i = i + 4
        elif code_array[i] == 2:
            j1 = code_array[i+1]
            j2 = code_array[i+2]
            j3 = code_array[i+3]
            code_array[j3] = code_array[j1] * code_array[j2]
            i = i + 4
        else:
            print('Cannot interpret intcode! Exiting.')
            break
    
    return code_array
```

```python
output_array = MakeCalculation(code_array)
```

## Part 2

```python
file = open('code.txt','r')
```

```python
code = file.read()
```

```python
code = code.replace('\n','')
```

```python
code_array = code.split(',')
```

```python
code_array = [int(x) for x in code_array]
```

```python
for i in range(100):
    for j in range(100):
        input_code = code_array.copy()
        input_code[1] = i
        input_code[2] = j
        output_code = MakeCalculation(input_code)
        print(f'noun={i}, verb={j}, output={output_code[0]}')        
```

```python
test_array = code_array.copy()
```

```python
test_array[1] = 64
test_array[2] = 21
MakeCalculation(test_array)[0]
```
